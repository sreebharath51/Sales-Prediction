```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("train2.csv")
x=data.drop(columns=['Item_Outlet_Sales'])
y=data['Item_Outlet_Sales']
data
```

```python
x=x.fillna(0)
#converting string into numeric using  frequency encoding method
#first we make a dictonary that maps each label to the count
#and we replace labels in the input data set x
Item_Identifier=x.Item_Identifier.value_counts().to_dict()
x.Item_Identifier=x.Item_Identifier.map(Item_Identifier)
Item_Fat_Content=x.Item_Fat_Content.value_counts().to_dict()
x.Item_Fat_Content=x.Item_Fat_Content.map(Item_Fat_Content)
Item_Type=x.Item_Type.value_counts().to_dict()
x.Item_Type=x.Item_Type.map(Item_Type)
Outlet_Identifier=x.Outlet_Identifier.value_counts().to_dict()
x.Outlet_Identifier=x.Outlet_Identifier.map(Outlet_Identifier)
Outlet_Size=x.Outlet_Size.value_counts().to_dict()
x.Outlet_Size=x.Outlet_Size.map(Outlet_Size)
Outlet_Location_Type=x.Outlet_Location_Type.value_counts().to_dict()
x.Outlet_Location_Type=x.Outlet_Location_Type.map(Outlet_Location_Type)
Outlet_Type=x.Outlet_Type.value_counts().to_dict()
x.Outlet_Type=x.Outlet_Type.map(Outlet_Type)
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

print('Mean of Target Variable : ',y_train.mean())

# predict the target on the test dataset
predict_train = [y_train.mean() for i in range(y_train.shape[0])]
 
# Root Mean Squared Error on training dataset
rmse_train = mean_squared_error(y_train,predict_train)**(0.5)
print('\nMSE on train dataset : ', rmse_train)
```

```python

x_train, x_val, y_train, y_val = train_test_split(x,y, test_size =0.3)

lreg= LinearRegression()
# training a linear regression model on train
lreg.fit(x_train,y_train)

# predicting on val
pred_val = lreg.predict(x_val)

# calculating mse
mse = np.mean((pred_val - y_val)**2)
mse
```

```python
lreg.score(x_val,y_val)
```
